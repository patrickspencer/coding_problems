# Leetcode 15. 3Sum

[Problem statement](https://leetcode.com/problems/3sum/)

```python
nums = [-1,0,1,2,-1,-4]
```

```python
n = len(nums)
target = 100
ans = []

def threesum(nums, target=0):
    nums.sort()
    n = len(nums)
    ans = []
    def sumfromi(i):
        l, r = i+1, n-1
        while l < r:
            s = nums[i] + nums[l] + nums[r]
            if s == target:
                ans.append((nums[i], nums[l], nums[r]))
                l += 1
                r -= 1
                while l < r and nums[l] == nums[l-1]:
                    l += 1
                while l < r and nums[r] == nums[r+1]:
                    r -= 1
            elif s < target:
                l += 1
            elif s > target:
                r -= 1

    for i in range(n):
        if i == 0 or nums[i] != nums[i-1]:
            if nums[i] > 0:
                break
            else:
                sumfromi(i)

    return ans
```

```python
nums = [-1,0,1,2,-1,-4]
threesum(nums)
```
