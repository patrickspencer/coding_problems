```python
def find_min_idx(a):
    n = len(a)
    l, r = 0, n-1
    while l < r:
        m = l + (r - l) // 2
        if a[m] <= a[m+1]:
            r = m
        else:
            l = m+1
    return l
```

```python
nums = [2, 3, 4, 5, 6, 7]
min_idx = find_min_idx(nums)
min_idx
```

```python
def bsearch(s, e, nums, target):
    l, r = s, e
    i = 0
    while l < r:
        m = l + (r - l) // 2
        if nums[m] == target:
            return m
        if nums[m] >= target:
            r = m
        else:
            l = m + 1
        i += 1
    
    return -1
```

```python
nums = [2, 3, 4, 5, 6, 7]
n = len(nums)
bsearch(0, n-1, nums, 5)
```
